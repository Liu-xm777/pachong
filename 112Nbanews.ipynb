```python
datas = []
def Get_news():
    cfs=soup.find(id="news_list_body").find_all('dt')
    for cf in cfs:
        title = cf.find('h3').text
        source =cf.find('a').get('href')
        content = cf.find('p').text
        pic = cf.find('span')
        pics=str(pic)
        picss=pics[73:235]
        img_url=picss
        print(title,source,content,img_url)
        datas.append([title,source,content,img_url])
    
Get_news()
```

```python
datas
```

```python
import pandas as pd
pd.DataFrame(datas)
df = pd.DataFrame(datas,columns=['標題','來源','摘要','圖片連結'])
df
```
